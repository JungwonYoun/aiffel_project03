# Feature 01 'strange_amt'
: 회원마다 카드 사용 금액의 평균과 표준편차를 구해서 해당 거래액이 평소 결제 금액보다 월등히 많거나 적은지 확인할 수 있는 변수 만들기

```python
# 카드번호별 group by
# 거래 금액의 평균과 표준편차 구하기
cc_amt_group = pd.DataFrame(cc_df.groupby('cc_num')['amt'].describe())
print(cc_amt_group)

# z-score 구해서 기존 데이터와 merge
cc_df['amt_z'] = cc_df.groupby('cc_num')['amt'].transform(lambda x : stats.zscore(x,ddof=1))
cc_df
```

# Feature 02 'Time'
:결제 시간을 카테고리화하여 변수 만들기

```python
cc_df['dt_time'] = pd.to_datetime(cc_df['trans_date_trans_time']).dt.hour
cc_df
```

```python
def time_categorize(hour):
    if 6 <= hour < 12 :
        return 'morning'
    elif 12 <= hour < 17 :
        return 'afternoon'
    elif 17 <= hour < 22 :
        return 'evening'
    else :
        return 'night'
```

```python
cc_df['time'] = cc_df['dt_time'].apply(lambda x : time_categorize(x))
cc_df.drop(['dt_time', 'trans_date_trans_time'], axis=1, inplace = True)
cc_df
```

```python
cc_df.groupby('time')['trans_num'].count()
```

# Feature 03 'Age'
:생년월일을 계산하여 나이 변수 만들기

```python
def calculate_age(birthdate_str):
    current_date = datetime.today()
    birthdate = datetime.strptime(birthdate_str, "%Y-%m-%d")
    
    
    # Calculate the age by subtracting the birthdate from the current date
    return current_date.year - birthdate.year - ((current_date.month,
              current_date.day) < (birthdate.month, birthdate.day))
```

```python
cc_df['age'] = cc_df['dob'].apply(lambda x: calculate_age(x))
cc_df
```

```python
# 불필요한 컬럼 제거
cc_df.drop('dob', axis=1, inplace = True)
```

```python
cc_df.groupby('age')['trans_num'].count()
```
